```python
from datetime import datetime, timezone, timedelta
from tabulate import tabulate
import csv

total = []

def acoes() -> dict:
    acoes = {1 : 'Tarefas Pendentes', 2 : 'Tarefas Concluídas', 
           3 : 'Cadastrar Tarefa', 4 : 'Concluir Tarefa',
           5 : 'Sair'
  }
    return acoes

def menu_inicial():
  menu = acoes()
  print('Olá, veja suas tarefas!\n')
  for chave in menu.keys():
    print(f' {chave} - {menu[chave]}')
  ask = int_menu('O que deseja fazer? Digite o número da ação ')
  return ask

def tarefas() -> list:
  lista = [['Data do cadastro', 'Descrição', 'Status', 'Data Limite'], 
           ['22/09/2022','Arrumar a casa', 'Pendente', '22/09/2022'], 
           ['22/09/2022','Ler livro', 'Pendente', '22/09/2022'], 
           ['22/09/2022','Lavar louça', 'Pendente', '22/09/2022'],
           ['22/09/2022','Lavar roupa', 'Pendente', '22/09/2022']]
  return lista

def get_data() -> str:

  diferenca = timedelta(hours=-3)
  fuso = timezone(diferenca)
  now = datetime.now().astimezone(fuso)
  data = now.strftime('%d/%m/%Y')
  return data

def nova_tarefa() -> list: 
  global total #preciso que o retorno dessa função apareça em todas as outras sem ter que aparecer os inputs, por isso disse que era global
  data = get_data()
  total = tarefas()
  nt_temp = [data, input("Descreva a atividade: "), 
             'Pendente', input("Digite a data limite: ")]
  total.append(nt_temp)
  r = str_sn("Deseja cadastrar mais uma atividade? [S/N] ")
  while r in 'Ss':
    nt_temp = [data, input("Descreva a atividade: "), 
             'Pendente', input("Digite a data limite: ")]
    total.append(nt_temp)
    r = str_sn("Deseja cadastrar mais uma atividade? [S/N] ")
    if r in 'Nn':
      break
  exportando_csv(total)  #quando add um tarefa, ele sobrescreve lá meu csv
  return total

def tarefas_pendentes():
  global total
  temp = []
  pendentes = [['Data do cadastro', 'Descrição', 'Status', 'Data Limite']]
  for i in total:
    temp.append(i[:])

  for i in temp:
    if i[2] == 'Pendente':
      pendentes.append(i)
  print(tabulate(pendentes, headers='firstrow'))

def tarefas_concluidas():
  global total
  temp = []
  concluidas = [['Data do cadastro', 'Descrição', 'Status', 'Data Limite']]
  for i in total:
    temp.append(i[:]) #só quero fazer cópias, não desejo mudar minha global com os ids.

  for i in temp:
    if i[2] == 'Concluida':
      concluidas.append(i)
  print(tabulate(concluidas, headers='firstrow'))

def concluir_tarefas() -> list:
  global total
  cont = 0
  temp = []

  for i in total:
    temp.append(i[:])

  temp[0].insert(0,'ID') #os ids so aparecem aqui pra orientar o usuário nesse momento e nada mais

  for i in temp[1:]:
    cont += 1
    i.insert(0,cont)

      
  print(tabulate(temp, headers='firstrow'))

  a = int_menu('Digite o ID da atividade que deseja concluir')

  total[a][2] = 'Concluida'

  print('='*60)
  print(f'A atividade -{total[a][1]}- foi concluída com sucesso!')
  print('='*60)

  r = str_sn("Deseja concluir mais uma atividade? [S/N] ")
  while r in 'Ss':

    a = int_menu('Digite o ID da atividade que deseja concluir')

    total[a][2] = 'Concluida'

    print('='*60)  #só pra deixar bonitinho
    print(f'A atividade -{total[a][1]}- foi concluída com sucesso!')
    print('='*60) 

    r = str_sn("Deseja concluir mais uma atividade? [S/N] ")
    if r in 'Nn':
      break
  exportando_csv(total) #quando att um tarefa, ele sobrescreve lá meu csv
  return total

def int_menu(msg:str) -> int: #função pra tratar erro nos inputs de inteiros
  while True:
    try:
      ask = int(input(msg))
    except (ValueError, TypeError):
      print('Valor inválido. Por favor digite um número válido')
      continue
    except (KeyboardInterrupt): 
      print('Erro! Interrompido pelo usuário')
      break
    else: 
      return ask
  
def str_sn(msg:str) -> str: #função pra tratar erro nos inputs de sim ou não
  dic = {'S' : 's', 's' : 's',
       'N' : 'n', 'n' : 'n'}
  while True:
    try:
      r = input(msg)
      print(dic[r])
    except (KeyError): #eu so preciso de S ou N e nada mais, se n tiver no meu dic, é erro de chave
      print('Valor inválido. Por favor, digite apenas [S/N]')
    else: 
      return r

def linhas(tam=60) -> str: #só pra deixar bonitinho
  return '-'*tam

def cabecalho(tit):  #só pra deixar bonitinho
  print(linhas())
  print(tit.center(60))
  print(linhas())

def exportando_csv(lista:list) -> csv:
    
    table = lista

    arquivo = open("table.csv","w")

    escritor = csv.writer(arquivo, delimiter=",", lineterminator="\n")

    escritor.writerows(table)

    arquivo.close()

def main(): #COMECE POR AQUI
  exportando_csv(tarefas()) 

  while True:
    tela = menu_inicial()

    if tela == 1:
      cabecalho('Tarefas Pendentes')
      tp = tarefas_pendentes()
      
    elif tela == 2:
      cabecalho('Tarefas Concluidas')
      tc = tarefas_concluidas()
      
    elif tela == 3:
      cabecalho('Cadastrar Tarefa')
      ct = nova_tarefa()
      
    elif tela == 4:
      cabecalho('Concluir Tarefa')
      cct = concluir_tarefas()
    
    elif tela == 5:
      print('Saindo do programa. Até logo!')  #só pra deixar bonitinho
      break
    else:
      print('Erro! Digite uma opção válida.') 
```

```python
main()
```
